```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df1_train = pd.read_csv("data/train_FD001.txt", sep=' ', header=None)
df1_test = pd.read_csv("data/test_FD001.txt", sep=' ', header=None)
df3_train = pd.read_csv("data/train_FD003.txt", sep=' ', header=None)
df3_test = pd.read_csv("data/test_FD003.txt", sep=' ', header=None)
```

```python
def data_merge(df1 ,df2) : 
    df2[0]= df2[0]+ df1[0].max()
    frames = [df1 , df2]
    df = pd.concat(frames)
    df.reset_index(level=0 , inplace=True)
    df.drop(['index'] , axis=1 , inplace=True)
    
    return df
```

```python
data_train.head()
```

```python
def data_cleaning(Data, t): 
    Data.drop([26,27], axis=1, inplace=True)
    column_name = ['ID','Cycle', 'op_set_1','op_set_2','op_set_3', 's1','s2','s3','s4','s5','s6','s7','s8','s9',
              's10','s11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21']
    Data.columns=column_name
    
    if (t=='train'):
        ## TTF calculation 

        failure_cycle = pd.DataFrame(Data.groupby('ID')['Cycle'].max())
        failure_cycle.reset_index(level=0, inplace=True)
        failure_cycle.columns=['ID','ttf']
        Data= pd.merge(Data, failure_cycle, on='ID')
        Data['ttf']= Data['ttf'] - Data['Cycle']

        ## labeling data 
        Data['labels'] = Data['ttf'].apply(lambda x: 0 if x <= 5 else 1 if x<=30 else 2 if x<=100 else 3)
        Data.drop(['ttf'], axis=1,inplace=True)
    
    return Data
```

```python
data_train = data_cleaning(data_merge(df1_train,df3_train), 'train')
```

```python
def data_validation(Data, RUL): 
    ## TTF calculation 
    max_cycle = pd.DataFrame(Data.groupby('ID')['Cycle'].max())
    max_cycle.reset_index(level=0, inplace=True)
    max_cycle.columns=['ID','max_cycle']
    Data= pd.merge(Data, RUL, on='ID')
    Data= pd.merge(Data, max_cycle, on='ID')
    Data['ttf']= Data['max_cycle'] - Data['Cycle'] + Data['ttf']

    ## labeling data 
    Data['labels'] = Data['ttf'].apply(lambda x: 0 if x <= 5 else 1 if x<=30 else 2 if x<=100 else 3)
    Data.drop(['ttf','max_cycle'], axis=1,inplace=True)
    
    return Data
```
